```python
#IMPORTS
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#LOAD & CLEAN DATA
df = pd.read_csv("nifty.csv")

df.columns = df.columns.str.lower()
df['date'] = pd.to_datetime(df['date'])
df = df.sort_values('date').reset_index(drop=True)

df.head()
```

```python
#SMART MONEY FEATURE

# Candle structure
df['body'] = abs(df['close'] - df['open'])
df['range'] = df['high'] - df['low']

# Volume spike
df['vol_ma'] = df['volume'].rolling(20).mean()
df['vol_spike'] = df['volume'] / df['vol_ma']

# Smart money metric
df['smart_money'] = (df['vol_spike'] * df['body']) / df['range']
```

```python
#ORIGINAL CHART

fig, ax1 = plt.subplots(figsize=(12,5))

# Price
ax1.plot(df['date'], df['close'], color='blue', label='Price')
ax1.set_ylabel("Price")

# Smart Money (second axis)
ax2 = ax1.twinx()
ax2.plot(df['date'], df['smart_money'], color='orange', label='Smart Money')
ax2.set_ylabel("Smart Money")

plt.title("Smart Money vs Price")
fig.legend(loc="upper left")
plt.show()
```

```python
df['price_smooth'] = df['close'].rolling(100).mean()
df['smart_smooth'] = df['smart_money'].rolling(100).mean()

fig, ax1 = plt.subplots(figsize=(12,5))
ax1.plot(df['date'], df['price_smooth'], color='blue', label='Price')
ax2 = ax1.twinx()
ax2.plot(df['date'], df['smart_smooth'], color='orange', label='Smart Money')

plt.title("Smart Money vs Price")
fig.legend(loc="upper left")
plt.show()
```
